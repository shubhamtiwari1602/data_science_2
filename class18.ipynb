```python
# t = np.linspace(-25,25,100)
t = np.arange(-2.0,2.0,0.1)
x= t 
y= t**2 
z= t**3 
fig = plt.figure(figsize= (4,4))
ax= fig.add_subplot(111,projection='3d')
ax.plot(x,y,z) 
```

```python
x= np.arange(-2,2,0.1) 
y= np.arange(-2,2,0.1) 
X,Y= np.meshgrid(x,y) 
Z= X**2+Y**2  
fig = plt.figure() 
ax= fig.add_subplot(111,projection='3d') 
ax.plot_surface(X,Y,Z) 
```

```python
cp= plt.contourf(X,Y,Z)
plt.clabel(cp,fontsize=10) # imp
plt.show()
```
